Abstraction in Object-Oriented Programming.</h5>

```python
## Inheritance

class Car:
    def __init__(self,color,max_speed,model):
        self.color=color
        self.max_speed=max_speed
        self.model=model

    def get_info(self):
        print(f"color is: {self.color}")
        print(f"max_speed is: {self.max_speed}")
        print(f"model is: {self.model}")

class Benz(Car):
    def __init__(self, color, max_speed, model,option):
        super().__init__( color, max_speed, model)
        self.option=option

    def get_all_info(self):
        self.get_info()
        print(f"option is :{self.option}")

c1=Benz("blue",300,2022,"hybrid")
c1.get_all_info()
```

```python
## Encapsulation

class Car:
    def __init__(self,color,max_speed,model):
        self.color=color
        self.max_speed=max_speed
        self.__model=model

    def get_info(self):
        print(f"color is: {self.color}")
        print(f"max_speed is: {self.max_speed}")
        print(f"model is: {self.__model}")


c1=Car("blue",300,2022)
c1._Car__model
```

```python
## Polymorphism

class Car:
    def __init__(self,color,max_speed,model):
        self.color=color
        self.max_speed=max_speed
        self.model=model

    def get_info(self):
        print(f"color is: {self.color}")
        print(f"max_speed is: {self.max_speed}")
        print(f"model is: {self.model}\n\n")

class Benz:
    def __init__(self,option):
        self.option=option

    def get_info(self):
        print(f"option of Benz is :{self.option}")

c1=Car("blue",300,2022)
b1=Benz("hybrid")
c1.get_info()
b1.get_info()
```

```python
## Abstraction


from abc import ABC, abstractmethod

class Car(ABC):
    def __init__(self,color,max_speed,model):
        self.color=color
        self.max_speed=max_speed
        self.model=model
        
    @abstractmethod
    def get_info(self):
        ...


class Benz(Car):
    def __init__(self, color, max_speed, model,option):
        super().__init__( color, max_speed, model)
        self.option=option

    def get_info(self):
        print(f"color is: {self.color}")
        print(f"max_speed is: {self.max_speed}")
        print(f"model is: {self.model}")
        print(f"option is :{self.option}")

c1=Benz("blue",300,2022,"hybrid")
c1.get_info()
```
